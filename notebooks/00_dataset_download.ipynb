# Download Project Data

This notebook uses the project's `src.data.download_data` module to fetch the required datasets.

```python
import os
import sys

# Add the src directory to system path so we can import our modules
sys.path.append(os.path.abspath("../src"))

from data.download_data import download_librispeech_sample
```

## Execute Download
This calls the function defined in `src/data/download_data.py`.

```python
DATA_DIR = "../data"

# This will download ~350MB if it doesn't already exist
dataset_path = download_librispeech_sample(DATA_DIR)

print(f"\nDataset is ready at: {dataset_path}")
```

## Verify Files

```python
import glob

# Check for FLAC files in the downloaded directory
files = glob.glob(os.path.join(dataset_path, "**", "*.flac"), recursive=True)

print(f"Found {len(files)} audio files.")
print("Sample files:", files[:3])
```
